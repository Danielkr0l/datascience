```python
import pandas as pd
import numpy as np
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
fig.show()
```

```python
import seaborn as sns
sns.set()
```

```python
plt.plot([0,1], [0,3], label = 'line', color = 'green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('tytuł')
_ = plt.legend()
```

```python
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin, 'g', label = 'sinus')
plt.plot(x,x_cos, 'r', label = 'cosinus')
_ = plt.legend()
```

Wiele wykresów

```python
plt.subplot(121)
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2],[1,5,2], 'r--')
```

```python
fig = plt.figure(figsize = (10,7))

plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[1,5,2], 'r--')
plt.subplot(133)
plt.plot([0,1,2],[1,4,5], 'ro')

fig.show()
```
