```python
df = pobieranie_danych('UBER')
df.info()
```

```python
df.head()
```

```python
df.describe().T
```

```python
df['Average'] = (df.Open + df.Close)/2
df.head()
```

```python
df = df.sort_index()
df.head()
```

```python
df[['Close']].shift(1)
```

```python
df['Close_shift'] = df.Close.shift(1)
df.head()
```

```python
df['Daily_change']= df.Close / df.Close_shift -1
df.head()
```
