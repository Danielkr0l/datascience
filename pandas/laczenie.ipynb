```python
import pandas as pd
```

```python
def wczytaj_dane(company = 'AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name = company, data_source = 'stooq')
apple = wczytaj_dane('AAPL')
amazon = wczytaj_dane('AMZN')
google = wczytaj_dane('GOOGL')
uber = wczytaj_dane('UBER')
```

```python
uber.info()
```

```python
apple.head()
```

```python
apple.columns = ['apple_'+col.lower() for col in apple.columns]
amazon.columns = ['amazon_'+col.lower() for col in amazon.columns]
google.columns = ['google_'+col.lower() for col in google.columns]
uber.columns = ['uber_'+col.lower() for col in uber.columns]
```

```python
uber.head()
```

```python
connect = pd.concat(objs=[apple,amazon,google,uber], axis = 1)
connect.head()
```

```python
pd.set_option('display.float_format', lambda x: f'{x: .2f}')
connect.describe().T
```

```python
connect.corr()
```

```python
closes = [col for col in connect.columns if col.endswith('close')]
closes
```
