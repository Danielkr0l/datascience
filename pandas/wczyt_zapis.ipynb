```python
import pandas as pd
import numpy as np
```

```python
def pobierz_dane(company='AMZN'):
  import pandas_datareader.data as web
  return web.DataReader(name = company, data_source = 'stooq')

df = pobierz_dane('META')
df.head()
```

**Zapis**

```python
df.to_csv('meta.csv')
```

```python
df_nov = df[(df.index.month == 2) & (df.index.year == 2023)]
df_nov
```

```python
df_nov.to_csv('luty.csv')
```
